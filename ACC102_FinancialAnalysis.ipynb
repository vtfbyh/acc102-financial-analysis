```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial']
```

```python
# Data Source: WRDS CSMAR, Data Acquired: 2026-04-12
data = {
     'Company': ['Midea','Gree','Haier']*5,
     'Year': [2020,2020,2020,
              2021,2021,2021,
              2022,2022,2022,
              2023,2023,2023,
              2024,2024,2024],
     'ROE': [22,23,18,21,22,17,20,22,16,21,23,17,22,24,18],
     'ProfitMargin': [8,15,5,7,14,5,8,16,6,9,17,6,10,18,6]
}

df = pd.DataFrame(data)
```

```python
# Preview the first 5 rows of the dataset
print("Data Preview:")
print(df.head())

# Check for missing values (data cleaning step)
print("\nMissing Value Check:")
print(df.isna().sum())
```

```python
# Caculate average ROE by company to compare profitability
print("Average ROE by Company (2020-2024):")
print(df.groupby('Company')['ROE'].mean().round(2))
```

```python
# Plot ROE trend comparison chart
df_pivot = df.pivot(index='Year', columns='Company', values='ROE')
df_pivot.plot(figsize=(10,5), marker='o', linewidth=2)
plt.title('ROE Comparison of Home Appliance Companies (2020-2024)', fontsize=14)
plt.ylabel('ROE (%)', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Business Insights & Conclusion
print("=== Key Business Insights (2020-2024) ===")
print("\n1. Profitability Performance:")
print("   - Gree Electric maintains the highest average ROE (22.8%) with strong stability,")
print("     demonstrating the strongest profitability among the three companies.")
print("   - Midea Group shows steady growth and consistent performance, with an average ROE of 21.2%.")
print("   - Haier has a relatively lower average ROE (17.2%) but remains stable, indicating solid operational fundamentals.")

print("\n2. Trend Analysis (from the chart):")
print("   - Gree's ROE saw a slight dip in 2021-2022 but rebounded strongly to 24% in 2024,")
print("     showing strong resilience and recovery capability.")
print("   - Midea's ROE remained stable throughout the period, with a steady upward trend after 2022.")
print("   - Haier's ROE hit a low in 2022 but has been recovering consistently since then.")

print("\n3. Investment Recommendation:")
print("   - For investors prioritizing stable high returns, Gree is the preferred choice.")
print("   - For those seeking balanced growth and risk, Midea offers a reliable option.")
print("   - Haier is suitable for investors focused on long-term recovery potential.")
```
